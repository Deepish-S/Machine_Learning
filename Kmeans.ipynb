```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

#make_blobs is a function provided by the scikit-learn library in 
#Python for generating synthetic datasets.
#It's commonly used for clustering and classification tasks.
#The function creates clusters of points, where each cluster 
#is characterized by a Gaussian distribution of points.
```

```python
X,y=make_blobs(n_samples=500,centers=4,cluster_std=.8,random_state=42)
```

```python
print("Shape of X:", X.shape)
```

```python
print("Shape of y:", y.shape)
```

```python
kmeans = KMeans(n_clusters = 4,  random_state = 42)
```

```python
kmeans.fit(X)
```

```python
labels=kmeans.labels_
labels
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X[:,0],X[:,1],c=labels,cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=100,c='red',label='centroids')
plt.title('K means Clustering')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
